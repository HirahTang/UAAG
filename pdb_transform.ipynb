```python
import os
from tqdm import tqdm
import pathlib
```

```python
pdb_folder_ls = os.listdir()
```

```python
pdb_folder_ls = [i for i in pdb_folder_ls if len(i) <=2]
```

```python
for pdb_folder in pdb_folder_ls:
    if not os.path.exists(pdb_folder):
        continue
    for pdb_ent in os.listdir(pdb_folder):
        pdb_name = pdb_ent.split('.')[0][-4:].upper()
        orig_path = os.path.join(pdb_folder, pdb_ent)
        if os.path.exists(orig_path):
            
            print(pdb_name)
            if os.path.exists('data/pdb/' + pdb_name + '.pdb'):
                print('PDB existed')
                continue
            else:
                os.rename(orig_path, 'data/pdb/' + pdb_name + '.pdb')       
            
            
        else:
            
            print(f'Path {orig_path} not found')
            continue
    pathlib.Path.rmdir(pathlib.Path(pdb_folder))
```
